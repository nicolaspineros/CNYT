```python
# Importaciones
from qiskit import QuantumCircuit #Clase para declarar circuitos
from qiskit import execute #Funcion para ejecutar circuitos cuanticos de manera simulada 
                            #o en computador cuantico real

from qiskit import Aer #Módulo donde podemos acceder al simulador qasm
from qiskit.visualization import plot_histogram, plot_bloch_vector # Función para graficar el output como diagrama de barras
%matplotlib inline
from math import sqrt, pi
# Esta último llamado me permite usar la librería mat´plotlib en el Notebook
```

**Ejercicio 1:** Cuando un qubit está inicializado en una superposición de los estados $|0\rangle$ y $|1\rangle$ (se logra aplicando la compuerta **H** al estado $|0\rangle$) dada por:

$$|\psi\rangle = \frac{1+i}{4} |0\rangle + \frac{-\sqrt{7}}{2\sqrt{2}} |1\rangle$$

El resultado de la medición debe ser o $|0\rangle$ con probabilidad del 12.5% o $|1\rangle$ con probabilidad del 87.5%. 

**a.**1 shot en un simulador.

```python
qc = QuantumCircuit(1)
backend = Aer.get_backend('statevector_simulator')
initial_state = [(1+1j)/4, -sqrt(7)/(2*sqrt(2))] #defenimos el estado inicial
qc.initialize(initial_state, 0)
result = execute(qc, backend, shots = 1).result()
counts = result.get_counts()
print(counts)
```

```python
plot_histogram(counts)
```

**Conclusion Ejercicio 1 a.**
efectivamente el resultado de la medición fue $|0\rangle$ con probabilidad del 12.5% o $|1\rangle$ con probabilidad del 87.5%. 1 shot en un simulador. teniendo de esta forma ningun tipo de ruido para el valor esperado en el simulador.

**b.**1000 shots en simulador

```python
qc = QuantumCircuit(1)
backend = Aer.get_backend('statevector_simulator')
initial_state = [(1+1j)/4, -sqrt(7)/(2*sqrt(2))] #defenimos el estado inicial
qc.initialize(initial_state, 0)
result = execute(qc, backend, shots = 1000).result()
counts = result.get_counts()
print(counts)
```

```python
plot_histogram(counts)
```

**Conclusion Ejercicio 1 a.**
efectivamente el resultado de la medición fue $|0\rangle$ con probabilidad del 12.5% o $|1\rangle$ con probabilidad del 87.5%. 

|      Resultados 0    |      Resultados 1     |
| :---------           |             --------: |
|       125/1000       |        875/1000       |
|         12,5%        |          87,5%        |

1000 shots en un simulador. teniendo de esta forma ningun tipo de ruido para el valor esperado en el simulador.

**c.**en un computador cuantico

```python
from qiskit import IBMQ
IBMQ.load_account()
```

```python
qc = QuantumCircuit(1)
initial_state = [(1+1j)/4, -sqrt(7)/(2*sqrt(2))] #defenimos el estado inicial
qc.initialize(initial_state, 0)
```

```python
proveedor = IBMQ.get_provider('ibm-q')
comp_cuantico = proveedor.get_backend('ibmq_london')
ejecucion = execute(qc, backend=comp_cuantico,shots=1000)

from qiskit.tools.monitor import job_monitor

job_monitor(ejecucion)

resultado = ejecucion.result()
conteos = resultado.get_counts()
print(conteos)
```

```python
plot_histogram(counts)
```

```python
from qiskit.visualization import plot_bloch_multivector
backend = Aer.get_backend('statevector_simulator')
out = execute(qc,backend).result().get_statevector()
plot_bloch_multivector(out)
```

```python
from qiskit import QuantumRegister, ClassicalRegister
from qiskit import BasicAer
from qiskit.tools.visualization import circuit_drawer

backend = BasicAer.get_backend('unitary_simulator')
```

```python
job = execute(qc, backend, shots = 1000)
conteos = job.result().get_unitary(qc, decimals = 3)
print(conteos)
```
