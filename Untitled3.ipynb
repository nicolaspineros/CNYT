```python
c1 = QuantumCircuit(1)
c1.h(0)
c1.x(0)
c1.z(0)
c1.h(0)
c1.draw(output='mpl')
```

```python
import numpy as np
```

```python
matriz_h = 1/np.sqrt(2)*np.array([[1, 1],[1, -1]])
matriz_x = np.array([[0,1],[1,0]])
matriz_z = np.array([[1,0],[0,-1]])
matriz_y = np.array([[0,-1.j],[1.j,0]])
```

```python
matriz_zy = np.matmul(matriz_z, matriz_x)
matriz_total = np.matmul(matriz_zy,matriz_y)
print(matriz_total)
```

```python
c2 = QuantumCircuit(3)
c2.h(2)
c2.x(0)
c2.z(1)
c2.draw(output='mpl')
```

```python
matriz_H_tensor_x = np.kron(matriz_z, matriz_h)
matriz_total_vertical = np.kron(matriz_H_tensor_x, matriz_x)
print(matriz_total_vertical)
```

```python
circuito = QuantumCircuit(2)
circuito.cx(0,1)
circuito.draw(output = 'mpl')
```

```python
from qiskit import *
%matplotlib inline
import numpy as np
from qiskit import Aer
```

```python
backend = Aer.get_backend('unitary_simulator')
result = execute(circuito, backend).result()

print(result.get_unitary(circuito, decimals=3))
```

```python
circuito_s = QuantumCircuit(2)
circuito_s.cx(0,1)
circuito_s.cx(1,0)
circuito_s.cx(0,1)
circuito_s.draw(output = 'mpl')
```

```python
backend = Aer.get_backend('unitary_simulator')
result = execute(circuito_s, backend).result()

print(result.get_unitary(circuito_s, decimals=3))
```
