```python
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram
%matplotlib inline
```

```python
oraculo = QuantumCircuit(6)
oraculo.cx(0, 3)
oraculo.cx(2, 3)
oraculo.x(3)
oraculo.cx(1, 4)
oraculo.cx(0, 5)
oraculo.cx(2, 5)
oraculo.ccx(5, 4, 3)
oraculo.ccx(3, 4, 5)
oraculo.ccx(3, 5, 4)
oraculo.ccx(3, 4, 5)
oraculo.draw(output = 'mpl')
```

```python
orac1 = oraculo.to_gate()
orac1.name = 'oraculo Simon'
```

```python
circuito = QuantumCircuit(6, 3)
circuito.h([0, 1, 2])
circuito.append(orac1, [0, 1, 2, 3, 4, 5])
circuito.h([0, 1, 2])
circuito.measure([0, 1, 2], [0, 1, 2])
circuito.draw(output='mpl')
```

```python
simulador = Aer.get_backend('qasm_simulator')
conteos = execute(circuito, backend=simulador, shots=1).result().get_counts()
print(conteos)
```

```python
plot_histogram(conteos)
```
